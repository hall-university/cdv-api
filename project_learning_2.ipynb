```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read data
df = pd.read_csv('winequalityN.csv')
```

```python
df.describe()
```

```python
# Check if dataset has None values
df.isna().sum()
```

```python
# Fill None values with mean value from category
df = df.fillna(df.mean())
```

```python
df.isna().sum()
```

```python
df.hist(bins=40, figsize=(10, 10))
plt.show()
```

```python
plt.figure(figsize=[18,7])
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
# Change categorical columns (type of wine) into binary column
df = pd.get_dummies(df, drop_first=True)
```

```python
df
```

```python
# Create binary quality (goo wine, bad wine)
df['binary_quality'] = [0 if x<7 else 1 for x in df.quality]

# Dependent features (all apart from quality and binary_quality)
x = df.drop(['quality', 'binary_quality'], axis=1)

# Independent features (only quality and binary_quality)
y = df['binary_quality']
```

```python
# Check how many good and bad wines are in the dataset
df['binary_quality'].value_counts()
```

```python
df
```

```python
# Standardizing data. We want to have data in the same range (standard). Now each column has it's own standard.
from sklearn.preprocessing import StandardScaler
x_features = x
x = StandardScaler().fit_transform(x)
```

```python
# Split data into test and train
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2)
```

```python
# Decision Tree model
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier

model1 = DecisionTreeClassifier()
model1.fit(x_train, y_train)
y1_pred = model1.predict(x_test)

print(classification_report(y_test, y1_pred))
```

```python
# Random Forest model
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier()
model2.fit(x_train, y_train)
y_pred2 = model2.predict(x_test)

print(classification_report(y_test, y_pred2))
```

```python
# Which features are the most important (RandomForest)
feat_importance = pd.Series(model2.feature_importances_, index=x_features.columns)
feat_importance.nlargest(25).plot(kind='barh', figsize=(10, 10))
```
